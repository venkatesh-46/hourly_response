```python
import pandas as pd
import numpy as np
```

```python
at=pd.read_excel('attendance_19.xlsx')
at=at[['Ids','TL Name']]
```

```python
agent_count=at.groupby('TL Name')['Ids'].count().to_frame()
agent_count.columns=['Agent Count']
```

```python
at=at.set_index('Ids')
```

```python
## 8 AM REPORT
valid_8=pd.read_csv('agent_hourly_8.csv')
valid_8=valid_8[['agent id','Valid response']]
valid_8['agent id']=pd.to_numeric(valid_8['agent id'],errors='coerce')
valid_8=valid_8.set_index('agent id')
valid_m_8=pd.merge(at,valid_8,left_index=True,right_index=True)
valid_m_8.reset_index(inplace=True)
valid_m_8=valid_m_8[['TL Name','Valid response']]
val_8=valid_m_8.groupby('TL Name')['Valid response'].sum()
val_8=val_8.to_frame()
mm_8=pd.concat([agent_count,val_8],axis=1)
mm_8['8:00 Avg']=mm_8['Valid response']/mm_8['Agent Count']
mm_8['8:00 Avg']=round(mm_8['8:00 Avg'],1)
mm_8=mm_8.replace(np.nan,0)
mm_8.columns=['Agent Count','08:00','08:00 Avg']
sum_agent_count_8=mm_8['Agent Count'].sum()
sum_8=mm_8['08:00'].sum()
avg_8=sum_8/sum_agent_count_8
avg_8=round(avg_8,1)
mm_8.loc["Grand Total"]=[sum_agent_count_8,sum_8,avg_8]
mm_8['Total Valid Forms']=mm_8['08:00']
mm_8['Avg Valid Forms']=mm_8['Total Valid Forms']/mm_8['Agent Count']
mm_8['Avg Valid Forms']=round(mm_8['Avg Valid Forms'],1)
mm_8=mm_8[['Agent Count','Total Valid Forms','Avg Valid Forms','08:00','08:00 Avg']]
mm_8.to_excel('8AM REPORT.xlsx')
```
